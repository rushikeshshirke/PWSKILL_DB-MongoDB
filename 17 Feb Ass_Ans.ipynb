```python
# # Q1.

# MongoDB is a popular NoSQL (non-relational) database management system that stores data in a flexible, 
# schema-less format.

# MongoDB is often preferred over SQL databases in the following scenarios:
#     When you need to store and query large amounts of unstructured or semi-structured data.
#     When you need a database that is highly scalable and can handle large amounts of traffic.
#     When you need a database that is flexible and can be easily adapted to changing requirements.
#     When you need a database that is easy to use and develop for.
```

```python
# # Q2.

# Document-oriented model: 
#     MongoDB stores data in JSON-like documents. This makes it easy to store and query complex data structures.
# Schema-less: 
#     MongoDB does not require a predefined schema. This makes it easy to add and modify data without having to make changes to the database structure.
# Scalability: 
#     MongoDB can be scaled horizontally to handle large amounts of data and traffic.
# Performance: 
#     MongoDB is designed for high performance, both read and write operations.
# Replication and high availability: 
#     MongoDB supports replication and sharding to achieve high availability and performance.
# Geo-spatial queries: 
#     MongoDB supports efficient geo-spatial queries, making it a good choice for applications that need to work with location data.
# Aggregation and analytics: 
#     MongoDB provides powerful aggregation and analytics capabilities, making it easy to perform complex data analysis.
```
